```python
#Simple bar chart showing revenues of major US tech companies
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos = np.arange(len(company))   #cant use str in bar chart so convert to nos
xpos
```

```python
plt.xticks(xpos,company)   #xpos nos to company names
plt.bar(xpos,revenue)
```

```python
plt.bar(xpos,revenue, label="Revenue")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
#Multiple Bars showing revenue and profit of major US tech companies
profit=[40,2,34,12]
#along with revenue and company we have profits
```

```python
plt.bar(xpos,revenue, label="Revenue")

plt.bar(xpos,profit,label="Profits")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
#since prev in same chart for diff chart +-0.2 to adjust xpos 
#numpy arr over list is used as it is easy for sub and creates a new arr with these value to avoid complications
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
#Horizontal bar chart using barh function
plt.barh(xpos,revenue, label="Revenue")

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technolog Stocks')
plt.legend()
```

```python
#In histograms, x axis contains a variable and y axis will be a frequency of that variable
#We have a sample data of blood sugar level of different patients, 
#we will try to plot number of patients by blood range and try to figure out how many patients are normal, 
#pre-diabetic and diabetic
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, rwidth=0.8) # by default number of bins(range) is set to 10
```

```python
#bins parameter
plt.hist(blood_sugar,rwidth=0.7,bins=4)  #reduce the bns and increase relative width gap between range
```

```python
#Histogram showing normal, pre-diabetic and diabetic patients distribution
#80-100: Normal
#100-125: Pre-diabetic
#80-100: Diabetic
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, color='g',histtype='step') #assigning own bins
```

```python
#Mutiple data samples in a histogram
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150],
         rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()
```

```python
#histtype=step
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,histtype='step')
```

```python
#horizontal orientation
plt.xlabel("Number Of Patients")
plt.ylabel("Sugar Level")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, orientation='horizontal')
```

```python
#Draw pie chart to track down home expenses
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

```python
#Draw pie chart as perfect circle
plt.pie(exp_vals,labels=exp_labels, shadow=True)
plt.axis("equal") #to get perfect circle
plt.show()  #to remove text
```

```python
#Show percentages for every pie. Specify radius to increase chart size
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5)  #autopct for displaying % 1.1 means 1decimal pt
plt.show()
```

```python
#Explode
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,4,0.1,0.2]) 
                                                                                        #explode comes out by that much %
plt.show()
```

```python
#counterclock and angle properties
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0,0,0.1,0.2],
        counterclock=True, startangle=45)
plt.show()                          #homerent starts at 45deg
```
